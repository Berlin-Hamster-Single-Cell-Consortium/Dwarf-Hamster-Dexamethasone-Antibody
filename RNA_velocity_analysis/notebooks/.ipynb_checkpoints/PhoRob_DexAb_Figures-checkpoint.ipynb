```python
df = pd.DataFrame(data=adata.obsm['X_diffmap'][:,1:4], index = adata.obs_names, columns=['DC1', 'DC2', 'DC3'])
df['condition'] = adata.obs.condition

x = adata.obsm['X_diffmap'][:, 1]
xmin = np.percentile(x, 5)
xmax = np.percentile(x, 95)

df = df[(x>xmin) & (x<xmax)]
df = df.sample(frac=1)  # shuffle rows
```

```python
vx, vy = adata.obsm['velocity_diffmap'][:, 1:3].T
x, y = adata.obsm['X_diffmap'][:, 1:3].T
sb=sns.jointplot(
    data=df,
    x="DC1", y="DC2", hue="condition", palette=adata.uns['condition_colors']
)
pl.savefig('figures/test2.pdf')
```

```python
x = adata.obsm['X_diffmap'][:, 1]
xmin = np.percentile(x, 1)
xmax = np.percentile(x, 99)
scv.pl.velocity_embedding_stream(adata[(x>xmin) & (x<xmax)].copy(), color='condition', basis='diffmap', figsize=[6,6], components='1,2', 
                                 legend_loc='right margin', alpha=0.2, size=300, min_mass=2, dpi=120, density=2, show=False)
pl.savefig('figures/test.pdf')
```

```python
vx, vy = adata.obsm['velocity_diffmap'][:, 1:3].T
x, y = adata.obsm['X_diffmap'][:, 1:3].T
```

```python
fig, ax = pl.subplots(1,1, figsize=[10,8])
ax.axis("off")

colors = adata.uns['condition_colors']
for group, c in zip(adata.obs['condition'].cat.categories, colors):
    mask = adata.obs['condition']==group
    # pl.scatter(x[mask], y[mask], alpha=0.2, color=c)
    pl.quiver(x[mask], y[mask], vx[mask], vy[mask], alpha=0.8, color=c, edgecolor='k', label=group)
ax.legend()
pl.show()
```
